## Ejercicio 1. Definición de la clase Cancion
1.1. Define  la clase Cancion, la cual va a tener 4 atributos básicos. (título,  artista, album, genero y duración)

1.2. Crea un método para la clase Cancion que nos muestre la información de la canción.

1.3. Crea un método que nos permita comparar la duración de dos canciones y nos indique cuál de las dos es más larga.

1.4. Crea un método que nos permita calcular la popularidad de una canción en función de varios criterios.

- La puntuación base debe ser de 50.
    
- Las canciones de entre 2 y 4 minutos de duración son más populares, por lo que suman 20 puntos. Las demás restan 10.  

- Las canciones de los géneros pop, rock, hip-hop y electrónicas son populares, por lo que suman 20 puntos. El resto de géneros restan 5 puntos.

- Si una canción pertenece a un álbum suma 10 puntos.

1.5. Crea dos instancias de la clase canción con la siguiente info: ("Stairway to Heaven", "Led Zeppelin", "Led Zeppelin IV", 8.02, "Rock") y ("What's Going On", "Marvin Gaye", "What's Going On", 3.53, "Soul")

1.6. Comprueba el funcionamiento de cada uno de los métodos de la clase



````python
class Cancion:
    #defino los atributos básicos
    def __init__(self, titulo, artista, album, duracion, genero):
        self.titulo = titulo
        self.artista = artista
        self.album = album
        self.genero= genero
        self.duracion = duracion

    #Creo el método para que nos de la información de la canción
    def info_completa(self):
        print(f"el tiítulo de la canción es: {self.titulo}")
        print(f"el artista es: {self.artista}")
        print(f"el album en el que está la canción es: {self.album}")
        print(f"su género es: {self.genero}")
        print(f"y tiene una duración de {self.duracion} minutos")

    #Creo el método que compara la duración de las dos canciones
    def compara_duración(self,cancion_2):
        cancion_2 = Cancion(cancion_2.titulo, cancion_2.artista, cancion_2.album, cancion_2.duracion, cancion_2.genero)
        if self.duracion < cancion_2.duracion:
            print(f"la canción {cancion_2.titulo} tiene mayor duración")
        else:
            print(f"la canción {self.titulo} tiene mayor duración")
    
    #Creo el método para calcular la popularidad:
    def popularidad(self):
        puntuación = 50
        #pop segun duración
        if float(self.duracion) < 4.00 and float(self.duracion) > 2.00:
            puntuación = puntuación + 20
        else:
            puntuación = puntuación - 10
        #pop segun género
        if self.genero == "electronica" or self.genero == "rock" or self.genero == "hip-hop" or self.genero == "pop":
            puntuación = puntuación + 20
        else:
            puntuación = puntuación - 5
        #pop segun album
        if self.album == None or self.album == "":
            puntuación = puntuación
        else:
            puntuación = puntuación + 10 

        print(f"la popularidad de la canción {self.titulo} es de {puntuación} sobre 100")

lunes = Cancion("Lunes","Estopa","Lunes",3.50, "pop")
hamburguesas = Cancion("hamburguesas", "Carolina Durante", None,1.00, "indie")
````

## Ejercicio 2. Definición de la subclase Colaboracion

````python
#Defino la nueva subclase y sus atributos
class Colaboracion (Cancion):
    def __init__(self, titulo, artista, album, duracion, genero, colaboradores):
        super().__init__(titulo, artista, album, duracion, genero)
        self.colaboradores = colaboradores
    #2.2 creo el método para que no de todos los detalles de una colaboración
    def datos_colabo(self):
        print(f"el tiítulo de la canción es: {self.titulo}")
        print(f"el artista es: {self.artista}")
        print(f"el album en el que está la canción es: {self.album}")
        print(f"su género es: {self.genero}")
        print(f"tiene una duración de {self.duracion} minutos")
        print(f"y los artistas colaboradores son {self.colaboradores}")
    #2.3 vamos a hacer que compruebe si es o no colaboración
    def colabo_o_cancion(self):
        if self.colaboradores == None or self.colaboradores == "" or str(type(self)) == "__main__.Cancion":
            print("es una canción sin colaboración")
        else:
            print(f"es una colaboración de {self.artista} con {self.colaboradores}")
    #2.4 para obtener el número de colaboradores contaremos las palabras completas que hay en la lista de colaboradores aunque esto puede dar error si un colaborador tiene un nombre con varias palabras
    def numero_colaboradores(self):
        contador = 0
        if self.colaboradores == None or self.colaboradores == "":
            contador = 0
        else:
            cadena_colaboradores = str(self.colaboradores)
            cadena_colaboradores = list(cadena_colaboradores.split(" "))
            for n in cadena_colaboradores:
                contador += 1
        print(f"esta canción tiene un total de {contador} colaboradores")
    #uso el código anterior actualizándolo
    def popularidad(self):
            puntuación = 50
            #pop segun duración
            if float(self.duracion) < 4.00 and float(self.duracion) > 2.00:
                puntuación = puntuación + 20
            else:
                puntuación = puntuación - 10
            #pop segun género
            if self.genero == "electronica" or self.genero == "rock" or self.genero == "hip-hop" or self.genero == "pop":
                puntuación = puntuación + 20
            else:
                puntuación = puntuación - 5
            #pop segun album
            if self.album == None or self.album == "":
                puntuación = puntuación
            else:
                puntuación = puntuación + 10 
            #Actualización con la colabo
            if self.colaboradores == None or self.colaboradores == "":
                puntuación = puntuación
            else:
                if len(self.colaboradores) > 0:
                    cadena_colaboradores = str(self.colaboradores)
                    cadena_colaboradores = list(cadena_colaboradores.split(" "))
                    contador = 0
                    for n in cadena_colaboradores:
                        contador += 1
                    puntuación = contador * 5 + puntuación 
            return f"la canción tiene una popularidad de {puntuación} sobre 100"
    #2.6 colaborador principal
    def colaborador_principal(self):
        if  self.colaboradores == None or self.colaboradores == "":  
            return "No hay colaboradores"
        else:
            cadena_colaboradores = str(self.colaboradores)
            cadena_colaboradores = list(cadena_colaboradores.split(" "))   
            return f"el colaborador principal es {cadena_colaboradores[0]}" 
        
                
                  
````

````python
#3.1 creo la subclase en función de la clase Canción
class CancionPremiada (Cancion):
    def __init__(self, titulo, artista, album, duracion, genero, premios):
        super().__init__(titulo, artista, album, duracion, genero,)
        self.premios = premios
    #3.2 uso el código de canción para esto mismo y le añado los premios
    def detalles_cancion_premiada(self):
        print(f"el tiítulo de la canción es: {self.titulo}")
        print(f"el artista es: {self.artista}")
        print(f"el album en el que está la canción es: {self.album}")
        print(f"su género es: {self.genero}")
        print(f"y tiene una duración de {self.duracion} minutos")
        print(f"ha sido premiada con {str(self.premios)}")
    #3.3 convierto los premios en lista ya que vienen dados en tipo lista en el enunciado y mido su longitud para sacar el número
    def num_premios(self):
        listado_premios = list(self.premios)
        return f"el número de premios es de {len(listado_premios)}"
    #3.4 premios importantes. mirare si en la lista premios hay alguno de los premios recibidos
    def premios_importantes(self):
        lista_premios_import = ["Grammy", "Oscar", "Golden Globe", "Brit Award"]
        listado_premios = list(self.premios)
        for n in listado_premios:
            if n in lista_premios_import:
                return True 
            continue
        else:
            return False
       
````

````python
class playlist:
    def __init__(self, nombre, canciones):
        self.nombre = nombre
        self.canciones = []
    #Creo el método para añadir canciones
    def anade_cancion(self, cancion_añadida):
        if type(cancion_añadida) == __main__.Cancion:
            self.canciones.append(cancion_añadida)
        else:
            print("la canción debe ser de la clase Cancion")
    #4.3 eliminar canciones de la playlist
    def elimino_cancion(self, cancion_eliminada):
        if type(cancion_eliminada) == __main__.Cancion:
            self.canciones.remove(cancion_eliminada)
    #4.4 mostrar la info
    def mostrar_info(self):
        print(f"el nombre de la playlist es {self.nombre}")
        print(f"las canciones de la playlist son")
        for n in self.canciones:
            print(f"{n}")
    #4.5 mostrar las canciones de un artista
    def canciones_artista(self,artista):
        for n in self.canciones:
            canciones_de_ese_cantante=[]
            if n.artista == artista:
                canciones_de_ese_cantante.append(n)
    #4.6 total canciones y minutos
    def conteo_canciones_minutos(self):
        num_canciones = 0
        num_minutos = 0
        for n in self.canciones:
            num_canciones = num_canciones + 1
            num_minutos = num_minutos + n.duracion
    #4.7 para ordenar en aleatorio busco en internet como hacerlp
    def random(self):
        from random import shuffle
        shuffle(self.canciones)
````
